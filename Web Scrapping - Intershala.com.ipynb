```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://internshala.com/fresher-jobs'
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup =BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
scrap_job_title = soup.find_all('div',class_='heading_4_5 profile')
scrap_job_title
```

```python
# Parsing Job Title
Title = []

for t in scrap_job_title:
    t = t.get_text().replace('\n','')
    Title.append(t)
Title
```

```python
len(Title)
```

```python
scrap_co_name = soup.find_all('div',class_='heading_6 company_name')
scrap_co_name
```

```python
# Parsing Company name
company = []

for c in scrap_co_name:
    company.append(c.text.strip())
    
company
```

```python
len(company)
```

```python
scrap_ctc = soup.find_all('div',class_='item_body')
scrap_ctc
                          
```

```python
# Parsing CTC
CTC = []

for t in scrap_ctc:
    CTC.append(t.text.strip())
CTC

CTC_new = []
for r in range(1,len(CTC),3):
    CTC_new.append(CTC[r])
CTC_new
```

```python
len(CTC)
```
